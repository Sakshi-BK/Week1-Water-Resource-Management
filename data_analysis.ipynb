```python
import pandas as pd
df = pd.read_csv("rainfall_in_india_1901-2015.csv")
df.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = df.dropna()
df["ANNUAL_RAINFALL"] = df.iloc[:, 2:14].sum(axis=1)
annual = df.groupby("YEAR")["ANNUAL_RAINFALL"].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=annual, x="YEAR", y="ANNUAL_RAINFALL")
plt.title("Average Annual Rainfall Over the Years")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
